```python
def unite_humans_and_examples(df):
    df = df.loc[df['data.Q0']!= 'GRAMMAR']
    answer = pd.DataFrame(columns=[])
    for user in np.unique(df['user_id']):
        temp_humans = df.loc[df['user_id'] == user]
        #print(np.sort(temp_humans['regex']))
        #print(temp_humans.shape)
        #print(temp_humans['data_id'].values[0])
        temp_answers = pd.read_json(f'../listener-experiments/data_{(temp_humans["data_id"].values[0])}.json')
        #temp_answers['regex'] = [REGEX_LOOK_UP_3[i] for i in temp_answers['regex']]
        #print(np.sort(temp_answers['regex']))

        temp_humans = temp_humans.merge(temp_answers, on='regex')
        #print(temp_humans)
        #print(temp_humans.shape)
        answer = pd.concat([answer, temp_humans])
    return answer
```

```python
def get_correct_percent(df, spec = False):
    df['method'] = [' '.join(i.split(' ')[0:int(len(i.split(' '))/2)]) for i in df['regex']]
    df['true'] = [' '.join(i.split(' ')[int(len(i.split(' '))/2): len(i.split(' '))]) for i in df['regex']]
    if (spec == True ):
        df['correct'] = [False for i in range(df.shape[0])]
    else:
        df['correct'] = [(df['method'][i] in REGEX_LOOK_UP_2[df['true'][i]] 
                    or df['method'][i] == df['true'][i]) for i,k in enumerate(df['method'])]
    print('Number Timed out:' , df.loc[df['true']== 'TIME'].index)
    print('Number Memory out:' ,df.loc[df['true']== 'MEMORY'].index)
    print('Correct vs wrong:', df['correct'].sum()/df.shape[0])
```
